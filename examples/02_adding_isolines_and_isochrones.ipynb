# Adding isolines/isochrones

## 1. generating isolines based on 500, 1500, 2500, 3500 ft distances

```python
isolines =il.OsmIsolines('Prospect Park, Brooklyn,NYC, USA',
                                 network_type = 'walk',
                                 values=[500, 1500, 2500, 3500],
                                 unit= 'ft',
                                 address_type='all',
                                 sample = 600)
```

```python
isolines.plot_isolines()
```

## 2. add isolines to an existing instance

We can add more isolines to an existing isoliner isnstance.
Here we add 1000, 2000 and 3000 and 4000 ft isolines
<ul>
<li>
The 4000 ft isoline could not be calculated because our instance was initiated using 3500 ft as an upper bound, however the 1000, 2000 and 3000 ft isolines were added
</li>


</ul>

```python
isolines.add_isolines([1000, 2000, 3000, 4000])
isolines.plot_isolines()
```
