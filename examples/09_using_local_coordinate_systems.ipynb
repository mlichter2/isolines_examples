```python
import isolines as il
import matplotlib.pyplot as plt
%matplotlib inline
from shapely.geometry import *
```

# Using local coordinate systems

note: not all coordinate systems will be reognized, global cartesian coordinate systems such as web mercator are not recognized.

## 1. Using a local coordinate system

```python
iso = il.OsmIsolines(Point(179370.985,664422.488),
                           values=[500, 250],
                          crs = 2039,
                    knn = 3)
iso.plot_isolines(plot_nodes=True, cmap = 'GnBu')
```

## 2. Using UTM

```python
iso = il.OsmIsolines(Point(561746,4155501),
                           values=[500],
                           address_type='point',
                          crs = 32610)
iso.plot_isolines(plot_nodes=True)
```
