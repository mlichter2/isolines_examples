```python
import isolines as il
import matplotlib.pyplot as plt
%matplotlib inline
from shapely.geometry import *
```

# Refining source location geometry

## 1. using a point location 
In this case the point location is located inside a building (Bozeman High School in Bozeman, Montana)
The point is snapped to the closest location on the network

```python
iso = il.OsmIsolines(Point(479093.366,160312.640),
                           values=[500],
                           address_type='point',
                          crs = 32100)
iso.plot_isolines(plot_nodes=True, plot_source_nodes=True,figsize = (10,10))
```

## 2. buffering the point location

```python
iso = il.OsmIsolines(Point(479093.366,160312.640).buffer(100),
                           values=[500],
                          crs = 32100)
iso.plot_isolines(plot_nodes=True, plot_source_nodes=True,figsize = (10,10))
```

## 3. Using a the actual building polygon extracted from OSM

```python
iso = il.OsmIsolines('Bozeman High School, Bozeman, Montana, USA',
                           values=[500],
                           sample = 200
                           )
iso.plot_isolines(plot_nodes=True, plot_source_nodes=True,figsize = (10,10))
```

```python
iso.change_isolines(knn = 15)
iso.plot_isolines(plot_nodes=True, plot_source_nodes=True,figsize = (10,10))
```
