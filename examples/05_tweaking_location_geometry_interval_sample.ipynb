# Tweaking  location geometry interval sample

## 1. walking isochrones from a polygon address with default sample intervals

Using the default ```address_type=  "all"``` - allows using
locations that are not points (polygons or lines to be isolined/isochroned)
<ul>
<li>
The geometry (polygon, multipolygon, linestring, multilinestring) will be sampled in evenly spaced intervals (the boundary of polygons or the line itself) and sample points from it will be added as nodes in the network, the default interval length is 100 m or 330 ft.
</li>
    
 <li>
Sample points from the geometry are shown in pink - we can see below that in this case, the default 330 ft sample might be too dense - we will change this in the next example
</li>

</ul>

````python
isochrones_from_poly = il.OsmIsolines('Prospect Park, Brooklyn,NYC, USA',
                                 network_type = 'walk',
                                 metric = 'time',
                                 values=[2.5, 5, 7.5],
                                 unit = 'ft',
                                 address_type='all')
isochrones_from_poly.plot_isolines(figsize = (10,10))
````

## 2. reducing the default polygon sample interval from 330 ft to 600 ft

Next we set the interval sample to 1000 ft instead of 300 ft


````python
isochrones_from_poly = il.OsmIsolines('Prospect Park, Brooklyn,NYC, USA',
                                 network_type = 'walk',
                                 metric = 'time',
                                 values=[2.5, 5, 7.5],
                                 unit= 'ft',
                                 address_type='all',
                                 sample = 1000)
isochrones_from_poly.plot_isolines(figsize = (10,10))
````
